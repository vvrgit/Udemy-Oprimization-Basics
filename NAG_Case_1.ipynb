```python
import pandas as pd
import random
```

```python
data=pd.read_excel("/content/SampleData.xlsx")
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.X,data.Y,test_size=0.2, random_state=42)
```

```python
Ns=X_train.shape[0]
gama=0.9
eta=0.1
epochs=1000
vm=0
vc=0
m=random.uniform(-1,1)
c=random.uniform(-1,1)
```

```python
for iter in range(epochs):
  for sample in range(Ns):
    gm=-(y_train.iloc[sample]-(m+gama*vm)*X_train.iloc[sample]-(c+gama*vc))*X_train.iloc[sample]
    gc=-(y_train.iloc[sample]-(m+gama*vm)*X_train.iloc[sample]-(c+gama*vc))
    vm=gama*vm-eta*gm
    vc=gama*vc-eta*gc
    m=m+vm
    c=c+vc
print("m=",round(m))
print("c=",round(c))
```

```python
#training MSE
y_train_pred=m*X_train+c
mse=((y_train-y_train_pred)**2).mean()
print("Training MSE=",round(mse))
```

```python
#Testing MSE
y_test_pred=m*X_test+c
mse=((y_test-y_test_pred)**2).mean()
print("Testing MSE=",round(mse))
```
