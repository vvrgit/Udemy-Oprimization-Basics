```python
 from scipy .optimize import linprog
```

```python
obj=[-50,-120]
lhs_ieq=[[7000,2000],[10,30]]
rhs_ieq=[700000,1200]
lhs_eq=[[1,1]]
rhs_eq=[100]
bnd=[(0,float("inf")),(0,float("inf"))]
opt=linprog(c=obj,A_ub=lhs_ieq,b_ub=rhs_ieq,A_eq=lhs_eq,b_eq=rhs_eq,bounds=bnd,method="simplex")
print("Wheat Land:",opt.x[0])
print("Barley Land:",opt.x[1])
```
