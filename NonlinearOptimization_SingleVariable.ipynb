```python
#step1: Intialization
import numpy as np
x=np.random.uniform(-10,10)
eta=0.1
epochs=100

for iter in range(epochs):
  #Step2: Calculate Gradient
  gx=2*x
  #Step3: Calculate step length
  delx=-eta*gx
  #Step4: update x
  x=x+delx

f=np.pow(x,2)+5
print("x=",round(x))
print("f=",f)
```
