```python
import pandas as pd
import numpy as np
import random
```

```python
data=pd.read_excel("/content/SampleData.xlsx")
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.X, data.Y, test_size=0.2, random_state=42)
```

```python
m=random.uniform(-10,10)
c=random.uniform(-10,10)
eta=0.1
epcohs=1000
```

```python
for iter in range(epcohs):
  for sample in range(X_train.shape[0]):
    gm=-(y_train.iloc[sample]-m*X_train.iloc[sample]-c)*X_train.iloc[sample]
    gc=-(y_train.iloc[sample]-m*X_train.iloc[sample]-c)
    delm=-eta*gm
    delc=-eta*gc
    m=m+delm
    c=c+delc
```

```python
print("m=",round(m))
print("c=",round(c))
```

```python
# Calculate predictions on the training data
y_train_pred = m * X_train + c

# Calculate the Mean Squared Error (MSE) on the training data
mse_train = ((y_train - y_train_pred) ** 2).mean()

print("Mean Squared Error on Training Data:", round(mse_train))
```

```python
y_test_pred=m*X_test+c
mse_test=((y_test-y_test_pred)**2).mean()
print("Mean Squared Error on Test Data:",round(mse_test))
```

```python
import numpy as np
import pandas as pd
import random
```

```python
data=pd.read_csv("/content/SGD-Assignment.csv")
data.head()
```

```python
data.describe()
```

```python
ph_min=6.7
ph_max=8.8
bico_min=35
bico_max=262
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize the MinMaxScaler
scaler = MinMaxScaler()
# Fit and transform the 'Ph' and 'Bicarbonates' columns
data[['Ph', 'Bicarbonates']] = scaler.fit_transform(data[['Ph', 'Bicarbonates']])
print("Normalized Data:")
display(data.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.Ph, data.Bicarbonates, test_size=0.2, random_state=42)
```

```python
m=random.uniform(-10,10)
c=random.uniform(-10,10)
eta=0.1
epochs=1000
```

```python
for iter in range(epochs):
  for sample in range(X_train.shape[0]):
    gm=-(y_train.iloc[sample]-(m*X_train.iloc[sample])-c)*X_train.iloc[sample]
    gc=-(y_train.iloc[sample]-(m*X_train.iloc[sample])-c)
    delm=-eta*gm
    delc=-eta*gc
    m=m+delm
    c=c+delc
print("Model Parameters: m=",round(m),"c=",round(c))
```

```python
print("Model Parameters: m=",m,"c=",c)
```

```python
# Calculate predictions on the training data
y_train_pred = m * X_train + c

# Calculate the Mean Squared Error (MSE) on the training data
mse_train = ((y_train - y_train_pred) ** 2).mean()

print("Mean Squared Error on Training Data:", mse_train)
```

```python
# Calculate predictions on the training data
y_test_pred = m * X_test + c

# Calculate the Mean Squared Error (MSE) on the training data
mse_test = ((y_test - y_test_pred) ** 2).mean()

print("Mean Squared Error on Tetsing Data:", mse_test)
```
