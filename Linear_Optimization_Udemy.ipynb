```python
from scipy.optimize import linprog
obj=[-6,-5]
lhs_ieq=[[1,1],[3,2]]
rhs_ieq=[5,12]
bnd=[(0,float("inf")),(0,float("inf"))]
opt=linprog(c=obj,A_ub=lhs_ieq,b_ub=rhs_ieq,bounds=bnd,method="simplex")
opt.x
print("No. of A:",round(opt.x[0]))
print("No. of B:",round(opt.x[1]))
```
