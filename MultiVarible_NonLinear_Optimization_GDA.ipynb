```python
#Step1: Initialization
import numpy as np
x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
eta=0.1
epochs=100

for iter in range(epochs):
  #Step2: Gradient Calculation with respect to each decision variable
  gx=6*x
  gy=10*y

  #Step3: Calculate step length for each decision varaibles
  delx=-eta*gx
  dely=-eta*gy

  #step4: Update all the decision variable
  x=x+delx
  y=y+dely

f=3*np.pow(x,2)+5*np.pow(y,2)+10
print("x=",round(x))
print("y=",round(y))
print("f=",f)
```
