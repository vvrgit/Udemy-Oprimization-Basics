```python
import numpy as np
import pandas as pd
import random
```

```python
data=pd.read_excel("/content/SampleData.xlsx")
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(data.X,data.Y,test_size=0.2,random_state=10)
```

```python
Ns=X_train.shape[0]
print("Ns=",Ns)
```

```python
eta=0.9
epochs=1000
m=random.uniform(-1,1)
c=random.uniform(-1,1)
```

image.png

```python
for iter in range(epochs):
  Gm=0
  Gc=0
  for sample in range(Ns):
    Gm=Gm+(y_train.iloc[sample]-m*X_train.iloc[sample]-c)*X_train.iloc[sample]
    Gc=Gc+(y_train.iloc[sample]-m*X_train.iloc[sample]-c)
  gm=-Gm/Ns
  gc=-Gc/Ns
  delm=-eta*gm
  delc=-eta*gc
  m=m+delm
  c=c+delc
print("model parameter m=",m)
print("model parameter c=",c)
```

```python
# Calculate predictions on the training data
y_train_pred = m * X_train + c

# Calculate the Mean Squared Error (MSE) on the training data
mse_train = ((y_train - y_train_pred) ** 2).mean()

print("Mean Squared Error on Training Data:", round(mse_train))
```

```python
# Calculate predictions on the testing data
y_test_pred = m * X_test + c

# Calculate the Mean Squared Error (MSE) on the training data
mse_test = ((y_test - y_test_pred) ** 2).mean()

print("Mean Squared Error on Testing Data:", round(mse_test))
```
