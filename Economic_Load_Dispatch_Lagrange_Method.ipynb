```python
import numpy as np
cost=np.array([[500,5.3,0.004],[400,5.5,0.006],[200,5.8,0.009]])
PD=800
sum1=0
sum2=0
pgen=[0,0,0]

for gen in range(3):
  sum1=sum1+cost[gen,1]/(2*cost[gen,2])
  sum2=sum2+1/(2*cost[gen,2])

lamda=(PD+sum1)/sum2

for gen in range(3):
  pgen[gen]=(lamda-cost[gen,1])/(2*cost[gen,2])
  print("Generation of unit",gen+1,"=",round(pgen[gen]))
```
